# Importando as bibliotecas 

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
```

# Carregando a base de dados

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

Imprimindo o X_train observamos que temos um cubo que contém várias imagens.

# Normalizando as Imagens

Para normalizar as imagens da base de treinamento e teste, dividimos cada pixel das imagens pelo maior valor que é 255. Com isso, os pixels estarão na faixa entre 0 e 1. Fazendo isso ganhamos velocidade no treinamento da rede neural.

```python
X_train = X_train/255.0

X_test = X_test/255.0
```

```python
X_train[0]
```

# Remodelando.

Vamos mudar a dimensão das bases de dados para ficarem no formato de um vetor, pois estamos trabalhando com uma rde neural densa.

```python
X_train.shape
```

```python
X_train = X_train.reshape(-1, 28*28)
X_train.shape
```

```python
X_train[0]
```

```python
X_test = X_test.reshape(-1, 28*28)
X_test.shape
```

# Construindo a Rede Neural

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=128, activation='relu', input_shape=(784,)))
model.add(tf.keras.layers.Dropout(0.2))

model.add(tf.keras.layers.Dense(units=10, activation='softmax'))
```

Compilando o modelo

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
```

```python
model.summary()
```

# Treinando o Modelo

```python
model.fit(X_train, y_train, epochs=10)
```

# Avaliação do Modelo e previsão

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
```

```python
test_loss, test_accuracy = model.evaluate(X_train, y_train)
```
